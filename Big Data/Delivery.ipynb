```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import isnan
from pandas import read_csv
```

```python
# Data import
order_recipe1 = pd.read_excel('ORDER_RECIPE1.xls')

order_recipe2 = pd.read_excel('ORDER_RECIPE2.xls')

order_recipe3 = pd.read_excel('ORDER_RECIPE3.xls')

order_recipe4 = pd.read_excel('ORDER_RECIPE4.xls')

order_recipe5 = pd.read_excel('ORDER_RECIPE5.xls')

order_recipe6 = pd.read_excel('ORDER_RECIPE6.xls')

order_recipe7 = pd.read_excel('ORDER_RECIPE7.xls')

order_recipe8 = pd.read_excel('ORDER_RECIPE8.xls')

orders = pd.read_excel('Orders.xls')

order_tank_inv = pd.read_excel('ORDERS_TANK_INV.xls')

compartment = pd.read_excel('COMPARTMENT.xls')

tank1 = pd.read_excel('TANK1.xlsx')

tractor = pd.read_excel('TRACTOR.xls')

trailer = pd.read_excel('TRAILER.xls')

# Columns to keep
columns_to_keep_order_recipe = ['ORDER_ID','PRODUCT_GROUP_CODE','ORDERED_QUANTITY','TERMINAL_PRODUCT_NUMBER']

# Keep specific columns
order_recipe1 = order_recipe1[columns_to_keep_order_recipe]

# Keep specific columns
order_recipe2 = order_recipe2[columns_to_keep_order_recipe]

# Keep specific columns
order_recipe3 = order_recipe3[columns_to_keep_order_recipe]

# Keep specific columns
order_recipe4 = order_recipe4[columns_to_keep_order_recipe]

# Keep specific columns
order_recipe5 = order_recipe5[columns_to_keep_order_recipe]

# Keep specific columns
order_recipe6 = order_recipe6[columns_to_keep_order_recipe]

# Keep specific columns
order_recipe7 = order_recipe7[columns_to_keep_order_recipe]

# Keep specific columns
order_recipe8 = order_recipe8[columns_to_keep_order_recipe]

columns_to_keep_orders = ['ORDER_ID','TRACTOR_CARRIER_NUMBER','DESTINATION_NUMBER']

# Keep specific columns and drop 'END_LOAD_TIME' and 'START_LOAD_TIME'
orders = orders[columns_to_keep_orders]
```

```python
columns_to_keep_compartment = ['CONTAINER_CODE','COMPARTMENT_NUMBER','CAPACITY']

# Keep specific columns
compartment = compartment[columns_to_keep_compartment]

columns_to_keep_tank1 = ['TANK_CODE','NAME','SHELL_CAPACITY']

# Keep specific columns
tank1 = tank1[columns_to_keep_tank1]

columns_to_keep_TRACTOR=['TRACTOR_NUMBER']

# Keep specific columns
tractor = tractor[columns_to_keep_TRACTOR]

columns_to_keep_TRAILER=['TRAILER_CODE','USABLE_CAPACITY']

# Keep specific columns
trailer = trailer[columns_to_keep_TRAILER]
```

```python
# Read all six Excel files into separate DataFrames
order_recipe_merge = pd.concat([order_recipe1, order_recipe2, order_recipe3, order_recipe4, order_recipe5, order_recipe6,order_recipe7,order_recipe8], ignore_index=True)
```

```python
Order_delivery = pd.merge(orders,order_recipe_merge, on='ORDER_ID', how='inner')
Order_delivery
```

```python
TrailerCompartmentMerge = pd.merge(trailer, compartment, left_on='TRAILER_CODE', right_on='CONTAINER_CODE', how='inner')
TrailerCompartmentMerge
```

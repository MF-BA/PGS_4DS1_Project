```python
def missing_percentage(df):
    total = df.isnull().sum().sort_values(ascending = False)
    percent = round(df.isnull().sum().sort_values(ascending = False)/len(df)*100,2)
    return pd.concat([total, percent], axis=1, keys=['Total','Percent'])
```

```python
missing_percentage(df)
```

```python
def drop_columns_with_full_missing(df):
    # Calculate missing percentages
    missing_percentages = missing_percentage(df)
    
    # Get columns with 100% missing values
    columns_to_drop = missing_percentages[missing_percentages['Percent'] == 100.0].index
    
    # Drop columns with 100% missing values
    df.drop(columns=columns_to_drop, inplace=True)
```

```python
def drop_columns_with_all_zeros(df):
    # Check if all values in each column are 0
    zero_columns = (df == 0).all()
    
    # Get the column names where all values are 0
    columns_to_drop = zero_columns[zero_columns].index
    
    # Drop columns where all values are 0
    df.drop(columns=columns_to_drop, inplace=True)
```

```python
def check_same_values(df, column1, column2):
    """
    Check if two columns in a DataFrame have the same values.
    
    Parameters:
        df (DataFrame): The DataFrame containing the columns.
        column1 (str): The name of the first column.
        column2 (str): The name of the second column.
        
    Returns:
        bool: True if the columns have the same values, False otherwise.
    """
    # Get unique values in each column
    unique_values1 = set(df[column1])
    unique_values2 = set(df[column2])
    
    # Check if the sets of unique values are equal
    if unique_values1 == unique_values2:
        return True
    else:
        return False
```

```python
def display_columns_with_same_values(df):
    for column in df.columns:
        if df[column].nunique() == 1:
            print(f"Column '{column}' has the same unique value: {df[column].unique()}")
```
